```python
def fill_na_with_nearest_avg(df, column_name, window=2):
    # Get a list of valid indices for iteration
    valid_indices = df.index.tolist()  # Use df.index to get actual indices

    for idx in valid_indices[window: len(valid_indices) - window]:  # Iterate through valid indices

        if pd.isna(df.loc[idx, column_name]):
            # Get the nearest 4 rows (2 before and 2 after)
            # Adjust to use .iloc to select rows based on position
            nearest_values = df.loc[valid_indices[valid_indices.index(idx) - window: valid_indices.index(idx) + window + 1], column_name]

            nearest_values = nearest_values.dropna()  # Drop NaNs to avoid errors in calculation
            if len(nearest_values) >= 4:  # Ensure there are at least 4 valid values
                avg_value = nearest_values.mean()
                df.loc[idx, column_name] = avg_value
    return df
```

```python
#WARNING, THIS CODE DOES TAKE ~30 MINUTES TO RUN, BUT SAVES >10000 ROWS, BE PREPARED TO DO SOMETHING ELSE
len_before = 0
len_after = 1
x = 0
count = 0

while (len_before != len_after) or (count < 5):
  df_cleaned = df_cleaned.sort_values(by='HINCP', ascending=False)
  len_before = len(df_cleaned[df_cleaned.isna().any(axis=1)])

  print(f"Iteration {x+1}: ")
  print(len_before)

  df_cleaned = fill_na_with_nearest_avg(df_cleaned, 'BDSP')
  df_cleaned = fill_na_with_nearest_avg(df_cleaned, 'NP')
  df_cleaned = fill_na_with_nearest_avg(df_cleaned, 'NRC')
  df_cleaned = fill_na_with_nearest_avg(df_cleaned, 'VALP')

  x += 1
  len_after = len(df_cleaned[df_cleaned.isna().any(axis=1)])

  if len_before == len_after:
    count += 1
  else:
    count = 0




#df_cleaned.reset_index(drop = True, inplace = False)

df_cleaned.head()
```

```python
def fill_na_with_majority(df, column_name, window=2):
  for idx in df.index[window: len(df) - window]:  # Iterate using .index[window:] to avoid index out of range
     if pd.isna(df.loc[idx, column_name]):
        # Get the 4 nearest rows (2 before and 2 after)
        # Use .iloc to get rows by position, ensuring it works with various index types
        # Convert idx to integer for iloc if it's not already
        idx_int = int(idx) if not isinstance(idx, int) else idx

        nearest_values = df.iloc[idx_int - window:idx_int + window + 1][column_name]
        nearest_values = nearest_values.dropna()  # Drop NaNs to avoid errors in counting

        if len(nearest_values) >= 4:  # Ensure there are at least 4 valid values
            # Calculate the majority (0 or 1)
            majority_value = nearest_values.mode()[0]  # mode returns the most frequent value
            df.loc[idx, column_name] = majority_value

  return df
```

```python
columns = list(df_cleaned.columns)
columns_to_fill = columns[6:]

print(columns_to_fill)

len_before = 0
len_after = 1
x = 0

while len_before != len_after:
  df_cleaned = df_cleaned.sort_values(by='HINCP', ascending=False)
  len_before = len(df_cleaned[df_cleaned.isna().any(axis=1)])

  print(f"Iteration {x+1}: ")
  print(len_before)
  for column in columns_to_fill:
    df_cleaned = fill_na_with_majority(df_cleaned, column)

  x += 1
  len_after = len(df_cleaned[df_cleaned.isna().any(axis=1)])

df_cleaned.head()
```
